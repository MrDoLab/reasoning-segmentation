# 2. 파라미터 설정


```python
# 파라미터 설정
TEST_MODE = True
NUM_TEST_SAMPLES = 10
BATCH_SIZE = 16
MAX_NEW_TOKENS = 256

GENERATION_CONFIG['max_new_tokens'] = MAX_NEW_TOKENS

VISUALIZE_RESULTS = True
NUM_VISUALIZE = 10
RANDOM_SAMPLES = True

print("Parameters configured:")
print(f"   - TEST_MODE: {TEST_MODE}")
print(f"   - NUM_TEST_SAMPLES: {NUM_TEST_SAMPLES}")
print(f"   - BATCH_SIZE: {BATCH_SIZE}")
print(f"   - MAX_NEW_TOKENS: {MAX_NEW_TOKENS}")
print(f"   - VISUALIZE_RESULTS: {VISUALIZE_RESULTS}")
print(f"   - NUM_VISUALIZE: {NUM_VISUALIZE}")
print(f"   - RANDOM_SAMPLES: {RANDOM_SAMPLES}")
```

# 3. 실행

```python
# 모델 로딩
model_manager = ModelManager()
model_manager.load_sam()
print("SAM loaded!")

print("\nLoading Qwen3-VL-8B...")
qwen3 = Qwen3Model()
qwen3.load()
print("Qwen3-VL-8B loaded!")
```

```python
# 이미지 로드
if TEST_MODE:
    test_df_subset = test_df.head(NUM_TEST_SAMPLES)
    print(f"TEST MODE: Processing only {NUM_TEST_SAMPLES} samples")
else:
    test_df_subset = test_df
    print(f"FULL MODE: Processing all {len(test_df)} samples")

all_data = []

for idx, row in tqdm(test_df_subset.iterrows(), total=len(test_df_subset), desc="Loading images"):
    image_path = Path(DATASET_DIR) / "test" / f"{row['ID']}.jpg"
    image = load_image(str(image_path))
    
    all_data.append({
        'id': row['ID'],
        'image': image,
        'query': row['query'],
        'mllm_pred': None,
        'sam_mask': None
    })

print(f"\nLoaded {len(all_data)} images")
print(f"Memory usage: ~{len(all_data) * 3 / 1024:.2f} GB")
```

```python
# MLLM + SAM Pipeline (Refinement)
if REFINEMENT_CONFIG['enabled']:
    print(f"Running Refinement Pipeline (max_rounds={REFINEMENT_CONFIG['max_rounds']})")
else:
    print("Running basic pipeline via run_refinement (refinement disabled)")

images = [item['image'] for item in all_data]
queries = [item['query'] for item in all_data]
image_ids = [item['id'] for item in all_data]

results = run_refinement(
    mllm_model=qwen3,
    sam_model=model_manager.sam,
    images=images,
    queries=queries,
    image_ids=image_ids,
    verbose=VERBOSE,
)

apply_refinement_results(all_data, results)

print(f"\nPipeline completed for {len(all_data)} images")
if REFINEMENT_CONFIG['enabled']:
    round_counts = {}
    for r in results:
        rounds = r['rounds']
        round_counts[rounds] = round_counts.get(rounds, 0) + 1
    print(f"   Round distribution: {round_counts}")
```

```python
# 제출 파일 생성
submission = []

for item in tqdm(all_data, desc="RLE encoding"):
    rle_string = rle_encode(item['sam_mask'])
    submission.append({
        'ID': item['id'],
        'Label': rle_string
    })

submission_df = pd.DataFrame(submission)

expected_ids = test_df_subset['ID'].tolist()
actual_ids = submission_df['ID'].tolist()
assert actual_ids == expected_ids, "Order mismatch!"

if TEST_MODE:
    submission_path = Path(OUTPUT_DIR) / "submission.csv"
else:
    submission_path = Path(OUTPUT_DIR) / SUBMISSION_FILE

submission_df.to_csv(submission_path, index=False)

print(f"\nSubmission saved: {submission_path}")
print(f"Total samples: {len(submission_df)}")
```

# 4. 테스트


```python
# 결과 시각화
if VISUALIZE_RESULTS and len(all_data) > 0:
    if RANDOM_SAMPLES:
        num_samples = min(NUM_VISUALIZE, len(all_data))
        sample_indices = np.random.choice(len(all_data), size=num_samples, replace=False)
        print(f"Visualizing {num_samples} random samples")
    else:
        sample_indices = list(range(min(NUM_VISUALIZE, len(all_data))))
        print(f"Visualizing first {len(sample_indices)} samples")
    
    # 1. 기본 시각화 (최종 결과)
    print("\n" + "="*80)
    print("📊 FINAL RESULTS (최종 결과)")
    print("="*80)
    for idx in sample_indices:
        visualize_result(all_data[idx])
    
    # 2. Refinement 진행 과정 시각화 (활성화된 경우만)
    if REFINEMENT_CONFIG['enabled']:
        print("\n" + "="*80)
        print("🔄 REFINEMENT PROGRESS (개선 과정)")
        print("="*80)
        
        # Round가 2 이상인 것만 선택 (개선이 일어난 케이스)
        refined_items = [(i, item) for i, item in enumerate(all_data) 
                         if item.get('refinement_rounds', 1) > 1]
        
        if refined_items:
            print(f"\n{len(refined_items)}개 이미지가 refinement를 거쳤습니다.")
            
            # 최대 5개만 상세 시각화
            show_items = refined_items[:5]
            print(f"상세 시각화: {len(show_items)}개")
            
            for i, item in show_items:
                if 'refinement_result' in item:
                    visualize_refinement_progress(item, item['refinement_result'])
        else:
            print("\n모든 이미지가 Round 1에서 통과했습니다! ✅")
            print("(Refinement가 필요 없었습니다)")
    else:
        print("\n(Refinement가 비활성화되어 진행 과정 없음)")
else:
    print("Visualization skipped (VISUALIZE_RESULTS=False or no data)")
```
